Making Histograms of Results. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv('allResults.txt', sep=',', header=None)
data.columns = ["KK", "RepRandPM", "HillClimbingPM", "SimAnnPM", "RepRandPre", "HillClimbingPre", "SimAnnPre"]
data.head()
```

```python
columns = ["KK", "RepRandPM", "HillClimbingPM", "SimAnnPM", "RepRandPre", "HillClimbingPre", "SimAnnPre"]
strings = {
            "KK" : "Karmarkar-Karp",
            "RepRandPM" : "Repeated Random for Plus/Minus Representation", 
            "HillClimbingPM" : "Hill Climbing for Plus/Minus Representation", 
            "SimAnnPM" : "Simulated Annealing for Plus/Minus Representation", 
            "RepRandPre" : "Repeated Random for Prepartitioned Representation", 
            "HillClimbingPre" : "Hill Climbing for Prepartitioned Representation", 
            "SimAnnPre" : "Simualted Annealing for Prepartitioned Reprsentation"
        }

fig, axs = plt.subplots(7, 1, figsize=(15, 50))

for alg, ax in zip(columns, axs.ravel()):
        ax.hist(data[alg], bins=20)
        ax.set(title="Results of " + strings[alg] + " Over 50 Problems", xlabel="Residues", ylabel="Count")

plt.show()
```

```python
data[['RepRandPre', 'HillClimbingPre', 'SimAnnPre']].plot(kind='box', showmeans=True, meanline=True, figsize=(10, 8))

plt.show()
```

```python
data[['KK']].plot(kind='box', showmeans=True, meanline=True, figsize=(8, 8))
plt.show()
```

```python
dataSameA = pd.read_csv('allResultsSameA.txt', sep=',', header=None)
dataSameA.columns = columns

newColumns = ["RepRandPM", "RepRandPre", "HillClimbingPM", "HillClimbingPre", "SimAnnPM", "SimAnnPre"]
strings = {
            "RepRandPM" : "Repeated Random for Plus/Minus Representation", 
            "HillClimbingPM" : "Hill Climbing for Plus/Minus Representation", 
            "SimAnnPM" : "Simulated Annealing for Plus/Minus Representation", 
            "RepRandPre" : "Repeated Random for Prepartitioned Representation", 
            "HillClimbingPre" : "Hill Climbing for Prepartitioned Representation", 
            "SimAnnPre" : "Simualted Annealing for Prepartitioned Reprsentation"
        }

fig, axs = plt.subplots(3, 2, figsize=(20, 30))

for alg, ax in zip(newColumns, axs.ravel()):
        ax.hist(dataSameA[alg], bins=20)
        ax.set(title="Results of " + strings[alg] + " on Same Problem", xlabel="Residues", ylabel="Count")

plt.show()
```

```python
dataSameA[['RepRandPre', 'HillClimbingPre', 'SimAnnPre']].plot(kind='box', showmeans=True, meanline=True, figsize=(10, 8))
dataSameA[['RepRandPM', 'HillClimbingPM', 'SimAnnPM']].plot(kind='box', showmeans=True, meanline=True, figsize=(10, 8))
dataSameA[['KK']].plot(kind='box', figsize=(8,8))
plt.show()
```
