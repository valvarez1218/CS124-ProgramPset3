Making Histograms of Results. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv('allResults.txt', sep=',', header=None)
data.columns = ["KK", "RepRandPM", "HillClimbingPM", "SimAnnPM", "RepRandPre", "HillClimbingPre", "SimAnnPre"]
data.head()
```

```python
columns = ["KK", "RepRandPM", "HillClimbingPM", "SimAnnPM", "RepRandPre", "HillClimbingPre", "SimAnnPre"]
strings = {
            "KK" : "Karmarkar-Karp",
            "RepRandPM" : "Repeated Random for Plus/Minus Representation", 
            "HillClimbingPM" : "Hill Climbing for Plus/Minus Representation", 
            "SimAnnPM" : "Simulated Annealing for Plus/Minus Representation", 
            "RepRandPre" : "Repeated Random for Prepartitioned Representation", 
            "HillClimbingPre" : "Hill Climbing for Prepartitioned Representation", 
            "SimAnnPre" : "Simualted Annealing for Prepartitioned Reprsentation"
        }

# fig, axs = plt.subplots(7, 1, figsize=(15, 50))

# for alg, ax in zip(columns, axs.ravel()):
#         ax.hist(data[alg], bins=20)
#         ax.set(title="Results of " + strings[alg] + " Over 50 Problems", xlabel="Residues", ylabel="Count")

# plt.show()
```

```python
data[['RepRandPre', 'HillClimbingPre', 'SimAnnPre']].plot(kind='box', showmeans=True, 
                                                          meanline=True, figsize=(10, 8),
                                                         ylabel='Residues', 
                                                        title='Results of Prepartition Algorithms Over 50 Problems')
plt.show()
```

```python
data[['KK']].plot(kind='box', showmeans=True, meanline=True, figsize=(8, 8), ylabel='Residues',
                 title='Results of Karmarkar-Karp On 50 Problems')
plt.show()
```

```python
data[['RepRandPM', 'HillClimbingPM', 'SimAnnPM']].plot(kind='box', showmeans=True, meanline=True,
                                                       figsize=(10, 8), ylabel='Residues (log scale)',
                                                      title='Results of Plus/Minus Algorithms Over 50 Problems')
plt.show()
```
